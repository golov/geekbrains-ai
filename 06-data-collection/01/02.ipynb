```python
import requests
import json
from pprint import pprint
```

```python
params = {
    'apiKey': 'API_KEY'
}

url = "https://api.spoonacular.com/recipes/random"
```

```python
response = requests.get(url, params=params)

j_data = response.json()

pprint(j_data)
```

```python
with open('02.json', 'w') as f:
    json.dump(j_data, f)
```
