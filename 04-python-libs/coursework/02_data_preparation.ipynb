```python
import numpy as np
import pandas as pd
```

```python
DATASET_PATH = './housing.csv'
PREPARED_DATASET_PATH = './housing_prepared.csv'
```

```python
df = pd.read_csv(DATASET_PATH, sep=',')
df.head(4)
```

```python
df.info()
```

```python
df.describe()
```

```python
# Обработка пропусков

df['housing_median_age_nan'] = 0
df.loc[df['housing_median_age'].isna(), 'housing_median_age_nan'] = 1
median = df['housing_median_age'].median()
df['housing_median_age'] = df['housing_median_age'].fillna(median)
```

```python
median_bedrooms = df['total_bedrooms'].median()
median_population = df['population'].median()
population_per_bedroom = median_population / median_bedrooms
population_per_bedroom
```

```python
df['total_bedrooms_nan'] = 0
df.loc[df['total_bedrooms'].isna(), 'total_bedrooms_nan'] = 1
df.loc[(pd.isnull(df['total_bedrooms'])), 'total_bedrooms'] = df['population'] * population_per_bedroom
```

```python
df.loc[df['total_bedrooms_nan'] == 1]
```

```python
df['population_nan'] = 0
df.loc[df['population'].isna(), 'population_nan'] = 1
df.loc[(pd.isnull(df['population'])), 'population'] = df['total_bedrooms'] / population_per_bedroom
```

```python
df.loc[df['total_bedrooms_nan'] == 1]
```

```python
median = df['total_bedrooms'].median()
df['total_bedrooms'].fillna(median, inplace=True)

median = df['population'].median()
df['population'].fillna(median, inplace=True)
```

```python
df.describe()
```

```python
df['ocean_proximity_nan'] = 0
df.loc[df['ocean_proximity'] == '-', 'ocean_proximity_nan'] = 1
```

```python
# Обработка выбросов

df['longitude_outlier'] = 0
df.loc[df['longitude'] >= 0, 'longitude_outlier'] = 1
df.loc[df['longitude'] > 0, 'longitude'] = df.loc[df['longitude'] > 0, 'longitude'] * -1
df.loc[df['longitude'] == 0, 'longitude'] = df['longitude'].median()
```

```python
df['latitude_outlier'] = 0
df.loc[(df['latitude'] <= 0) | (df['latitude'] > 50), 'latitude_outlier'] = 1
df.loc[(df['latitude'] <= 0) | (df['latitude'] > 50), 'latitude'] = df['latitude'].median()
```

```python
total_rooms_max_value = np.quantile(df['total_rooms'], q=0.975)
total_rooms_min_value = np.quantile(df['total_rooms'], q=0.025)
condition = (df['total_rooms'] > total_rooms_max_value) | (df['total_rooms'] < total_rooms_min_value)
df.loc[condition, 'total_rooms'] = df['total_rooms'].median()
```

```python
total_bedrooms_max_value = np.quantile(df['total_bedrooms'], q=0.975)
total_bedrooms_min_value = np.quantile(df['total_bedrooms'], q=0.025)
condition = (df['total_bedrooms'] > total_bedrooms_max_value) | (df['total_bedrooms'] < total_bedrooms_min_value)
df.loc[condition, 'total_bedrooms'] = df['total_bedrooms'].median()
```

```python
# Выкидываем не валидне данные

# Все дома со стоимостью выше 500к скопляются в одну точку. Это не верные данные.
df = df[df['median_house_value'] < 500000]
```

```python
df.describe()
```

```python
# Отбор признаков

df['population_per_room'] = df['population'] / df['total_rooms']
df = pd.concat([df, pd.get_dummies(df['ocean_proximity'])], axis=1)
df.drop(columns=['id', '-', 'ocean_proximity', 'ocean_proximity_nan', 'population_nan', 'total_bedrooms_nan'], inplace=True)
df.head()
```

```python
df.describe()
```

```python
df.to_csv(PREPARED_DATASET_PATH, index=False, encoding='utf-8', sep=';')
```
